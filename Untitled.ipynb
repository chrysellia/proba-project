```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import norm

# Donnes des tches
tasks = {
    'A': {'predecessors': ['F', 'E'], 'optimistic': 1, 'most_likely': 2, 'pessimistic': 3},
    'B': {'predecessors': ['A', 'G'], 'optimistic': 1, 'most_likely': 3, 'pessimistic': 5},
    'C': {'predecessors': ['B'],      'optimistic': 1, 'most_likely': 3, 'pessimistic': 4},
    'D': {'predecessors': ['E'],      'optimistic': 12, 'most_likely': 13, 'pessimistic': 14},
    'E': {'predecessors': ['J'],         'optimistic': 3, 'most_likely': 5, 'pessimistic': 8},
    'F': {'predecessors': ['I', 'J'],      'optimistic': 2, 'most_likely': 6, 'pessimistic': 8},
    'G': {'predecessors': ['H'], 'optimistic': 7, 'most_likely': 8, 'pessimistic': 10},
    'H': {'predecessors': ['I'],      'optimistic': 7, 'most_likely': 8, 'pessimistic': 10},
    'I': {'predecessors': ['L'],      'optimistic': 6, 'most_likely': 9, 'pessimistic': 12},
    'J': {'predecessors': ['L', 'K'],      'optimistic': 1, 'most_likely': 2, 'pessimistic': 3},
    'K': {'predecessors': [],      'optimistic': 3, 'most_likely': 6, 'pessimistic': 9},
    'L': {'predecessors': [],         'optimistic': 1, 'most_likely': 3, 'pessimistic': 5},
}

# Calcul de la dure moyenne, cart-type, variance
for t in tasks:
    o = tasks[t]['optimistic']
    m = tasks[t]['most_likely']
    p = tasks[t]['pessimistic']
    mean = (o + 4*m + p) / 6
    std_dev = (p - o) / 6
    tasks[t]['mean'] = mean
    tasks[t]['std_dev'] = std_dev
    tasks[t]['variance'] = std_dev**2

# Cration du graphe avec networkx
G = nx.DiGraph()

for t, data in tasks.items():
    for pred in data['predecessors']:
        G.add_edge(pred, t)

# Visualisation du graphe
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, font_size=12)
plt.title("Graphe du projet (PERT)")
plt.show()
```

```python
# 2. Chemin critique bas sur la dure moyenne
G_duration = nx.DiGraph()
for t in tasks:
    G_duration.add_node(t, mean=tasks[t]['mean'], std_dev=tasks[t]['std_dev'])
for t, data in tasks.items():
    for pred in data['predecessors']:
        G_duration.add_edge(pred, t, weight=tasks[pred]['mean'])

critical_path = nx.dag_longest_path(G_duration, weight='weight')
critical_duration = sum(tasks[t]['mean'] for t in critical_path)
critical_variance = sum(tasks[t]['variance'] for t in critical_path)
critical_std_dev = np.sqrt(critical_variance)

print("2. Dure moyenne du projet :", round(critical_duration, 2), "jours")
print("   Chemin critique :", "  ".join(critical_path))
```

```python
# 3. Probabilit de terminer en 32 jours
target_days = 32
Z = (target_days - critical_duration) / critical_std_dev
prob_32 = norm.cdf(Z)
print("3. Probabilit que le projet soit fini en 32 jours :", round(prob_32 * 100, 2), "%")
```

```python
# 4. Dure avec une probabilit de 98%
Z_98 = norm.ppf(0.98)
duration_98 = critical_duration + Z_98 * critical_std_dev
print("4. Dure avec 98% de probabilit :", round(duration_98, 2), "jours")
```

```python
# 5. Modification des dures de la tche H
tasks['H']['mean'] = 9
tasks['H']['std_dev'] = 5
tasks['H']['variance'] = 25

# Recalcul du chemin critique avec la nouvelle tche H
critical_duration_new = sum(tasks[t]['mean'] for t in critical_path)
critical_variance_new = sum(tasks[t]['variance'] for t in critical_path)
critical_std_dev_new = np.sqrt(critical_variance_new)

Z_new = (target_days - critical_duration_new) / critical_std_dev_new
prob_32_new = norm.cdf(Z_new)
duration_98_new = critical_duration_new + Z_98 * critical_std_dev_new

print("5. Aprs modification de H :")
print("   Nouvelle dure moyenne :", round(critical_duration_new, 2), "jours")
print("   Nouvelle probabilit de finir en 32 jours :", round(prob_32_new * 100, 2), "%")
print("   Nouvelle dure pour 98% :", round(duration_98_new, 2), "jours")
```
